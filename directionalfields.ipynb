```python
import numpy as np
import matplotlib.pyplot as plt
# plot boundary
L = 5
# number of grid points to plot in each direction
n = 20
# Matrix
A = np.array([ [5, -1],
[3, 1]])
xx = np.linspace(-L, L, num=n)
yy = np.linspace(-L, L, num=n)
x1, x2 = np.meshgrid(xx, yy)
# computing eigenvalues and eigenvectors
evals, evecs = np.linalg.eig(A)
print('eigs', evals)
print('evecs', evecs)
# computing Ax
u = A[0,0] * x1 + A[0,1] * x2
v = A[1,0] * x1 + A[1,1] * x2
# normalize u, v to have constant arrow length
norm = np.sqrt(u**2 + v**2)
norm[norm == 0] = 1
u /= norm
v /= norm
# plotting the direction field
plt.quiver(x1, x2, u, v)
# time points
tt = np.linspace(-2, 10, num=1000)
# extracting real and complex parts of eigenvalues and eigenvectors
lam = np.real(evals[0])
mu = np.imag(evals[0])
a = np.real(evecs[:,0])
b = np.imag(evecs[:,0])
# fundamental solutions
sol_1 = np.exp(lam * tt) * ( np.cos(mu * tt) * a.reshape(-1,1) - np.sin(mu * tt) *
b.reshape(-1,1) )
sol_2 = np.exp(lam * tt) * ( np.sin(mu * tt) * a.reshape(-1,1) + np.cos(mu * tt) *
b.reshape(-1,1) )
# plotting the trajectories of solutions for various c_1, c_2
for c_1 in range(-2, 2):
    for c_2 in range(-2,2):
        sol = c_1 * sol_1 + c_2 * sol_2
plt.plot(sol[0], sol[1])
# plot properties
plt.axis('scaled')
plt.grid()
plt.xlim(-L,L)
plt.ylim(-L,L)
plt.show()
```
